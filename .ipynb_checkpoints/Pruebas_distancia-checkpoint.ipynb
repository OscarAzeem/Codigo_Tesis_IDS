```python
from sklearn.neighbors import DistanceMetric

dist = DistanceMetric.get_metric('euclidean')
X = [[0, 1, 2],
         [3, 4, 5]]
dist.pairwise(X)
```

```python
import numpy as np  
import matplotlib.pyplot as plt  
import seaborn as sns
def graph(formula, x_range):  
    x = np.array(x_range)  
    y = eval(formula)
    plt.plot(x, y)  
    plt.show()
```

```python
graph('x**3+2*x-4', np.linspace(0.1, 1))
#numpy.linspace(start, stop, num=50, endpoint=True, retstep=False, dtype=None)[source]
```

```python
graph('(1/(1+(2.71828**(-0.1/x))))', np.linspace(0.1, 1))
```

```python
x=np.linspace(0.1,1)
formula='(1/(1+(2.71828**(-0.1/x))))'
x = np.array(x)  
y = eval(formula)
plt.plot(x, y)
plt.title('Función $sgm(d_{ist})$',fontsize=30)
plt.xlabel("Distancia",fontsize=22)
plt.ylabel("Fuerza",fontsize=22)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.show()
```
