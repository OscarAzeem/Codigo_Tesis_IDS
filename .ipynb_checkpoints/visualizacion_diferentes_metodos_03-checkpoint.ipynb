```python
#converting to numpy
```

```python
X=nsl_training.values
y=L_nsl_training_indexbinary.values.T
```

```python
def ploteo(X,y,dimensiones,title=None):
    """X, despues de haber sido aplicada la transformaci贸n. y, Labels actuales
    X,y-> arreglo numpy"""
    #normalizando la transformaci贸n (despues de ser normalizada)
    x_min,x_max=np.min(X,0),np.max(X,0)
    X=(X-x_min)/(x_max-x_min)
    plt.figure()
    ax=plt.subplot(111)
    final=X.shape[0]-1
    print('el valor final es: %d' %final)
    for i in list(range(0,final)):
        plt.text(X[i, 0], X[i, 1], str(y[0,i]),
                 color=plt.cm.Set1(y[0,i] / dimensiones),
                 fontdict={'weight': 'bold', 'size': 9})
    plt.xticks([]),plt.yticks([])
    if title is not None:
        plt.title(title)
    
```

```python
def ploteo2(X,y,dimensiones,title=None):
    """X, despues de haber sido aplicada la transformaci贸n. y, Labels actuales
    X,y-> arreglo numpy"""
    #normalizando la transformaci贸n (despues de ser normalizada)
    x_min,x_max=np.min(X,0),np.max(X,0)
    X=(X-x_min)/(x_max-x_min)
    plt.figure()
    ax=plt.subplot(111)
    final=X.shape[0]-1
    print('el valor final es: %d' %final)
    for i in list(range(0,final)):
        plt.text(X[i, 0], X[i, 1], str(y[i]),
                 color=plt.cm.Set1(y[i] / dimensiones),
                 fontdict={'weight': 'bold', 'size': 9})
    plt.xticks([]),plt.yticks([])
    if title is not None:
        plt.title(title)
```

# ---.---

```python
#selecting just a few rows
X=X[0:5000,:]
y=y[0:5000,:]
```

```python
print("Computing PCA projection")
t0 = time()
X_pca = decomposition.TruncatedSVD(n_components=2).fit_transform(X)
ploteo(X_pca,y,2,"Principal Components projection of the digits (time %.2fs)" % (time() - t0))
plt.show()
```

```python
#----------------------------------------------------------------------
# t-SNE embedding of the digits dataset
print("Computing t-SNE embedding")
tsne = manifold.TSNE(n_components=2, init='pca', random_state=0)
t0 = time()
X_tsne = tsne.fit_transform(X)

ploteo(X_tsne,y,2,"t-SNE embedding of the digits (time %.2fs)" % (time() - t0))

plt.show()
```
