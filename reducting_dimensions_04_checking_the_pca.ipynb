```python
import matplotlib.pyplot as plt
import pandas as pd

from sklearn.decomposition import PCA as sklearnPCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
from sklearn.datasets.samples_generator import make_blobs

from pandas.tools.plotting import parallel_coordinates
```

```python
plt.rcParams['figure.figsize']= 8,8
#muestra las figuras de matplot
import warnings
warnings.filterwarnings('ignore')
#para desactivar la regresi√≥n
```

```python
import seaborn as sns; sns.set(color_codes=True)
```

```python
import numpy as np; np.random.seed(8)
```

```python
#reducting dimentions
#Declaring PCA
pca = sklearnPCA(n_components=2) #2-dimensional PCA
```

```python
#Reading the datasets
test=pd.read_csv("./Final_Datasets/STD_FD_NSL_Test.csv")
tra=pd.read_csv("./Final_Datasets/STD_FD_NSL_Tra.csv")
```

```python
#Reading the labels
#Index binary
L_nsl_testing_indexbinary=pd.read_csv("./Final_Datasets/L_nsl_testing_indexbinary.csv")
#Index by Group
L_nsl_testing_indexbygroup=pd.read_csv("./Final_Datasets/L_nsl_testing_indexbygroup.csv")
#Index tipe of attack
L_nsl_testing_indextipeofattack=pd.read_csv("./Final_Datasets/L_nsl_testing_indextipeofattack.csv")


#nsl TRAINING DATASET
L_nsl_training_indexbinary=pd.read_csv("./Final_Datasets/L_nsl_training_indexbinary.csv")
#Index by Group
L_nsl_training_indexbygroup=pd.read_csv("./Final_Datasets/L_nsl_training_indexbygroup.csv")
#Index tipe of attack
L_nsl_training_indextipeofattack=pd.read_csv("./Final_Datasets/L_nsl_training_indextipeofattack.csv")
```

```python
L_nsl_testing_indexbinary=L_nsl_testing_indexbinary['IndexBinary']
```

```python
L_nsl_training_indexbinary=L_nsl_training_indexbinary['IndexBinary']
```

```python
L_nsl_testing_indexbygroup=L_nsl_testing_indexbygroup['IndexGroup']
```

```python
L_nsl_training_indexbygroup=L_nsl_training_indexbygroup['IndexGroup']
```

```python
#Gettin the PCA components

ttra=pd.DataFrame(pca.fit_transform(tra))
ttest=pd.DataFrame(pca.transform(test))

#ttest= pd.DataFrame(pca.fit_transform(test))
#ttra=pd.DataFrame(pca.fit_transform(tra))
```

```python
#Binary Classification
sns.set_style('whitegrid')

plt.scatter(ttest[L_nsl_testing_indexbinary==0][0], ttest[L_nsl_testing_indexbinary==0][1], label='No ataque', c='blue',
            marker='s')
plt.scatter(ttest[L_nsl_testing_indexbinary==1][0], ttest[L_nsl_testing_indexbinary==1][1], label='Ataque', c='red')

plt.legend(prop={'size':17},frameon=True,shadow=True)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.title('ENSLKDD. Clasificaci√≥n Binaria',fontsize=22)


#ax.yaxis.label.set_size(40)
plt.savefig("std_2d_ensl_2_classes.png", dpi=300)
plt.show()
```

```python
#Binary Classification
plt.scatter(ttra[L_nsl_training_indexbinary==0][0], ttra[L_nsl_training_indexbinary==0][1], label='No ataque', c='blue',
            marker='s')
plt.scatter(ttra[L_nsl_training_indexbinary==1][0], ttra[L_nsl_training_indexbinary==1][1], label='Ataque', c='red')

plt.legend(prop={'size':17},frameon=True,shadow=True)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.title('NSLKDD. Clasificaci√≥n Binaria',fontsize=22)

plt.savefig("std_2d_nsl_2_classes.png", dpi=300)
#ax.yaxis.label.set_size(40)

plt.show()
```

# FIVE classes

```python
#five classes
plt.scatter(ttest[L_nsl_testing_indexbygroup==1][0], ttest[L_nsl_testing_indexbygroup==1][1], 
            label='Dos', c='blue',marker='s')
plt.scatter(ttest[L_nsl_testing_indexbygroup==2][0], ttest[L_nsl_testing_indexbygroup==2][1], 
            label='Normal', c='red',marker='d')
plt.scatter(ttest[L_nsl_testing_indexbygroup==3][0], ttest[L_nsl_testing_indexbygroup==3][1], 
            label='Probe', c='sienna',marker='o')
plt.scatter(ttest[L_nsl_testing_indexbygroup==4][0], ttest[L_nsl_testing_indexbygroup==4][1], 
            label='R2L', c='green',marker='x')
plt.scatter(ttest[L_nsl_testing_indexbygroup==5][0], ttest[L_nsl_testing_indexbygroup==5][1], 
            label='U2R', c='turquoise',marker='v')

plt.legend(prop={'size':17},frameon=True,shadow=True)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.title('ENSLKDD. 5 Clases Principales',fontsize=22)
plt.savefig("std_2d_ensl_5_classes.png", dpi=300)

#ax.yaxis.label.set_size(40)

plt.show()
```

```python
#five classes
plt.scatter(ttra[L_nsl_training_indexbygroup==1][0], ttra[L_nsl_training_indexbygroup==1][1], 
            label='Dos', c='blue',marker='s')
plt.scatter(ttra[L_nsl_training_indexbygroup==2][0], ttra[L_nsl_training_indexbygroup==2][1], 
           label='Normal', c='red',marker='d')
plt.scatter(ttra[L_nsl_training_indexbygroup==3][0], ttra[L_nsl_training_indexbygroup==3][1], 
           label='Probe', c='sienna',marker='o')
plt.scatter(ttra[L_nsl_training_indexbygroup==4][0], ttra[L_nsl_training_indexbygroup==4][1], 
            label='R2L', c='green',marker='x')
plt.scatter(ttra[L_nsl_training_indexbygroup==5][0], ttra[L_nsl_training_indexbygroup==5][1], 
            label='U2R', c='turquoise',marker='v')

plt.legend(prop={'size':17},frameon=True,shadow=True)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.title('NSLKDD. 5 Clases Principales',fontsize=22)
plt.savefig("std_2d_nsl_5_classes.png", dpi=300)
plt.show()
```
