```python
import numpy as np  
import matplotlib.pyplot as plt  
import seaborn as sns
def graph(formula, x_range):  
    x = np.array(x_range)  
    y = eval(formula)
    plt.plot(x, y)  
    plt.show()
```

```python
graph('x**3+2*x-4', np.linspace(0.1, 1))
#numpy.linspace(start, stop, num=50, endpoint=True, retstep=False, dtype=None)[source]
```

```python
graph('(1/(1+(2.71828**(-0.1/x))))', np.linspace(0.1, 1))
```

```python
x=np.linspace(0.001,1)
formula='(1/(1+(2.71828**(-0.1/x))))'
x = np.array(x)  
y = eval(formula)
plt.plot(x, y)
plt.title('Funcin $sgm(d_{ist})$',fontsize=30)
plt.xlabel("Distancia",fontsize=22)
plt.ylabel("Fuerza",fontsize=22)
plt.xticks(fontsize=20)#,linespacing=.5)
#plt.xticks(x[::.5], y[::.5], fontsize=20)
#plt.xticks(x[::frequency], my_xticks[::frequency])
plt.yticks(fontsize=20)
plt.savefig("./sgm.png", dpi=300)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

x = [0,5,9,10,15]
y = [0,1,2,3,4]
fig, ax = plt.subplots()
ax.plot(x,y)
start, end = ax.get_xlim()
ax.xaxis.set_ticks(np.arange(start, end, 1))
ax.xaxis.set_major_formatter(ticker.FormatStrFormatter('%0.1f'))
plt.show()
```
